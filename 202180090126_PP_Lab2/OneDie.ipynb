```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Bar Plot

```python
rolls=[rd.randrange(1,7) for i in range(600000)]
```

```python
values, frequencies = np.unique(rolls, return_counts=True)
```

```python
values
```

```python
frequency
```

```python
title = f'Rolling a Six-Sided Die {len(rolls):,} Times'
```

```python
plt.figure(figsize=(10,6))
bar_chart=sns.barplot(x=values, y=frequency, palette='viridis')
bar_chart.set(title=title)
bar_chart.set(xlabel='Die Value', ylabel='Frequency')
bar_chart.set_ylim(top=max(frequencies) * 1.10)

for bar, frequency in zip(bar_chart.patches, frequencies):
    text_x = bar.get_x() + bar.get_width() / 2.0
    text_y = bar.get_height()
    text = f'{frequency:,}\n{frequency / len(rolls):.3%}'
    bar_chart.text(text_x, text_y, text, fontsize=11, ha='center', va='bottom')
plt.show()
```
