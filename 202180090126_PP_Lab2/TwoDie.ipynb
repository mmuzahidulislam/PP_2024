```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Bar Plot

```python
rolls1=[rd.randrange(1,7) for i in range(600)]
rolls2=[rd.randrange(1,7) for i in range(600)]
```

```python
rolls=np.array(rolls1) + np.array(rolls2)
```

```python
values, frequencies = np.unique(rolls, return_counts=True)
```

```python
title = f'Rolling Two Six-Sided Die for {len(rolls):,} Times'
```

```python
plt.figure(figsize=(10,6))
bar_chart=sns.barplot(y=values, x=frequencies, palette='bright', orient='h')
bar_chart.set(title=title)
bar_chart.set(ylabel='Die Value', xlabel='Frequency')
bar_chart.set_xlim(right=max(frequencies) * 1.10)

for bar, frequency in zip(bar_chart.patches, frequencies):
    text_x = bar.get_width()
    text_y = bar.get_y() + bar.get_height() / 2.0
    text = f'{frequency:,}\n{frequency / len(rolls):.3%}'
    bar_chart.text(text_x, text_y, text, fontsize=11, ha='left', va='center')
    
plt.show()
```
